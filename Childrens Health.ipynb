#### 3. DATA CLEANING

Dealing with missing values: Here we are goung to replace all the missing values in our dataset by getting the mean per counties and filling each missing values in the column by that mean value

```python
df1 = df1.copy()
columns_with_missing = df1.columns[df1.isnull().any()].tolist()

for col in columns_with_missing:
    df1[col] = df1[col].fillna(df1.groupby('period')[col].transform('mean'))
    
df1
```

converting the "period" column to Datetime format

```python
# Convert 'period' column to datetime format
df1['period'] = pd.to_datetime(df1['period'], format='%b-%y')

# Extract the year from 'period' and create a new 'year' column
df1['year'] = df1['period'].dt.year
```

```python
# Exctrating the month from period column and creating a new column "Month"
df1['Month'] = df['period'].str[:3]
```

```python
gdf.rename(columns={'Name':'county'}, inplace=True)
```

```python
gdf['county'] = gdf['county'] + ' County'
```

```python
df_geo = pd.merge(df1, gdf, on='county', how='left')
df_geo.head(5)
```

```python

# Create a GeoDataFrame from the DataFrame
gdf = gpd.GeoDataFrame(df_geo, geometry='geometry')

# List of columns to plot maps for
columns_to_plot = [
    'Total Dewormed', 'Acute Malnutrition',
    'stunted 6-23 months', 'stunted 0-<6 months', 'stunted 24-59 months',
    'diarrhoea cases', 'Underweight 0-<6 months', 'Underweight 6-23 months',
    'Underweight 24-59 Months'
]

# Group by county and calculate the sum for each column
df_grouped = gdf.groupby('county')[columns_to_plot].sum().reset_index()

# Merge the grouped data back to the GeoDataFrame
gdf_merged = gdf.merge(df_grouped, on='county', suffixes=('', '_sum'))

# Plot maps for each column
for column in columns_to_plot:
    plt.figure(figsize=(14, 8))
    ax = plt.gca()

    # Customize the map appearance (optional)
    vmin = gdf_merged[column + '_sum'].min()  # Minimum value for the colormap
    vmax = gdf_merged[column + '_sum'].max()  # Maximum value for the colormap
    gdf_merged.plot(column=column+'_sum', cmap='RdYlBu_r', edgecolor='black', linewidth=0.5, legend=True, ax=ax, vmin=vmin, vmax=vmax)

    # Set axis labels and title (you can modify these)
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.title(f'Map of {column} in Kenyan Counties')

    # Adjust plot layout
    plt.tight_layout()

    # Show the plot
    plt.show()
```

Plotting the numbers in the Kenyan Map according to counties, we can see how various health condition affect counties in general. This is the data the has been captured from 2021 to May 2023.

>1.Nairobi county has the highest number of children dewormed 2,000,000, compared to other counties in Kenya. One major explanation for this is because Nairobi has a bigger population in kenya and has more health facilities as compared to other counties, hence high number of dewormed children.

>2.Acute mulnutrition on the other hand is more prevelant in Wajir county, Over 30,000. This may be due to prolonged drought, farmine and insecurity in the region making it hard for families to frow food and feed their children healthy.

>3.Diarrhoea cases are more prevelant in Nairobi- 300,000. followed by Kilifi and Turkana counties around 225,000. Nakuru and Mandera come third (175,000) with Uasin Gishu and Kiambu following closely at around 150,000.
